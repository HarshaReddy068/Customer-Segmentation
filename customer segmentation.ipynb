```python
df.drop(["Customer ID"], axis=1, inplace=True)
df.head()
```

```python
plt.figure(1, figsize=(15, 6))
n = 0
for x in ['Age', 'Previous Purchases', 'Review Rating']:
    n += 1
    plt.subplot(1, 3, n)
    plt.subplots_adjust(hspace=0.5, wspace=0.5)
    sns.distplot(df[x], bins=20)
    plt.title('Distplot of {}'.format(x))
plt.show()
```

```python
plt.figure(1, figsize=(15, 6))
n = 0

# Assuming 'Payment Method' is a categorical variable
for a in ['Payment Method']:
    n += 1
    plt.subplot(1, 1, n)
    plt.subplots_adjust(hspace=0.5, wspace=0.5)
    sns.countplot(data=df, x=a)
    plt.title('Countplot of {}'.format(a))

plt.show()
```

```python
plt.figure(1, figsize=(15, 6))
n = 0

# Assuming 'Payment Method' is a categorical variable
for a in ['Season']:
    n += 1
    plt.subplot(1, 1, n)
    plt.subplots_adjust(hspace=0.5, wspace=0.5)
    sns.countplot(data=df, x=a)
    plt.title('Countplot of {}'.format(a))

plt.show()
```

```python
plt.figure(figsize=(15, 5))
sns.countplot(y='Gender', data=df)
plt.show()
```

```python
plt.figure(1, figsize=(15, 7))
n = 0
for cols in ['Age', 'Previous Purchases', 'Review Rating']:
    n += 1
    plt.subplot(1, 3, n)
    sns.set(style="whitegrid")
    plt.subplots_adjust(hspace=0.5, wspace=0.5)
    sns.violinplot(x=cols, y='Gender', data=df)
    plt.ylabel('Gender' if n == 1 else '')
    plt.title('violin plot')
plt.show()
```

```python
age_18_25 = df.Age[(df.Age >= 18) & (df.Age <= 25)]
age_26_35 = df.Age[(df.Age >= 26) & (df.Age <= 35)]
age_36_45 = df.Age[(df.Age >= 36) & (df.Age <= 45)]
age_46_55 = df.Age[(df.Age >= 46) & (df.Age <= 55)]
age_55above = df.Age[df.Age >= 56]

agex = ["18-25", "26-35", "36-45", "46-55", "55+"]
agey = [len(age_18_25.values), len(age_26_35.values), len(
    age_36_45.values), len(age_46_55.values), len(age_55above.values)]

plt.figure(figsize=(15, 6))
sns.barplot(x=agex, y=agey, palette="mako")
plt.title("Number of Customer and Ages")
plt.xlabel("Age")
plt.ylabel("Number of Customer")
plt.show()
```
